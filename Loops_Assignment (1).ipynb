## Class Attendance Tracker

### Goal:
    Track only present students using continue.

### Steps:

    Have a list of students:
    
    students = ["Ali", "Sara", "Ahmed", "Ayesha"]
    
    
    Loop through each student.
    
    Ask "Is <name> present (yes/no)?"
    
    If ‚Äúno‚Äù ‚Üí use continue to skip.
    
    Add only present students to a present_list.

### Example:

    Is Ali present? yes  
    Is Sara present? no  
    Is Ahmed present? yes  
    Is Ayesha present? yes  
    Present Students: ['Ali', 'Ahmed', 'Ayesha']

```python
students=[]
status=[]

while True:
    name = input("Enter student name : ")
    if name.lower() == "done":
        break

    present = input("Enter status (P for Present / A for Absent): ")
    
    students.append(name)
    status.append(present.upper())


print("Attendance Report:")
for i in range(len(students)):
    print(students[i], ":", status[i])


total_present = status.count("P")


print("Total Present:", total_present)

```

## üè® 8. Hotel Room Booking

### Goal:
    Manage simple room booking.

### Task:

    Create a list of rooms [101, 102, 103, 104, 105]
    
    Ask user to choose a room.
    
    If room already booked, print ‚ÄúRoom not available.‚Äù
    
    Continue until all rooms are booked.

### Example:

    Available rooms: [101, 102, 103, 104, 105]
    Choose room: 103
    Room 103 booked successfully!

```python
rooms = [101, 102, 103, 104, 105]

while rooms:  
    print("\nAvailable rooms:", rooms)
    
    choice = input("Choose a room to book: ")

    if not choice.isdigit(): 
        print("Please enter a valid room number.")
        continue

    choice = int(choice)
    if choice in rooms:
        rooms.remove(choice)
        print(f"Room {choice} booked successfully!")
    else:
        print("Room not available.")

print("\nAll rooms are booked!")
```

## üìÖ 9. Daily Task Reminder

### Goal:
    Create a daily to-do list.

### Task:

    Ask user to enter 5 tasks.
    
    Print all tasks using a loop.
    
    Then allow user to remove a completed task.

### Example:

    Enter task 1: Study
    Enter task 2: Gym
    Today's Tasks: ['Study', 'Gym', 'Read']
    Enter completed task: Gym
    Remaining Tasks: ['Study', 'Read']

```python
tasks=[]
for i in range(5):
    task = input(f"Enter task {i+1}: ")
    tasks.append(task)
print("\nYour To-Do List:")
for i, task in enumerate(tasks, start=1):
    print(f"{i}. {task}")
completed = input("\nEnter the task you completed: ")
if completed in tasks:
    tasks.remove(completed)
    print(f"\n'{completed}' has been removed from your list!")
else:
    print(f"\nTask '{completed}' not found in your list.")

print("\nUpdated To-Do List:")
for i, task in enumerate(tasks, start=1):
    print(f"{i}. {task}")
```

## Hotel Room Booking Tracker
### Goal:

    Record bookings for rooms on each floor.

### Steps:

    Outer loop ‚Üí for 3 floors.
    
    Inner loop ‚Üí for 4 rooms per floor.
    
    Ask user if booked (Y/N).
    
    Count total booked and available rooms.

    Floor 2:

## Invoice Tax Calculator
### Goal:

Take multiple items in multiple invoices and calculate tax per invoice.

### Steps:

Outer loop ‚Üí for 3 invoices.

Inner loop ‚Üí 4 items each.

Input item price, add to list.

Calculate total + 10% tax.

Print invoice total.

### Example:
    Invoice 1:
    Items: [100, 200, 300, 400]
    Invoice 2:
    Items: [250, 150, 350, 250]
    ‚Üí Totals: [1100, 1100]

```python
invoice_totals = []
for inv in range(1,4):
    print(f"Invoice {inv}:")
    items = []
    for i in range(1,5):
        price = float(input(f"Item {i} price: "))
        items.append(price)
    total = sum(items)
    total_with_tax = total * 1.10
    invoice_totals.append(total_with_tax)
    print(f"Invoice {inv} total (with 10% tax): {total_with_tax}")
print("Totals:",invoice_totals)
```

## Student Result Sheet
### Goal:

    Generate report cards for each student in a class.

### Steps:

    Outer loop ‚Üí 5 students.
    
    Inner loop ‚Üí 4 subjects.

    If marks < 33 ‚Üí print ‚ÄúFailed in Subject‚Äù.

    Calculate total and grade per student.

### Example:
    Student 1: [80, 90, 70, 85] ‚Üí Grade: A  
    Student 2: [50, 32, 45, 40] ‚Üí Failed in Subject 2

```python
for student in range(1,6):
    print(f"Student {student}:")
    marks = []
    failed = False
    for sub in range(1,5):
        mark = float(input(f"Subject {sub}: "))
        marks.append(mark)
        if mark < 33:
            print(f"Failed in Subject {sub}")
            failed = True
    total = sum(marks)
    if not failed:
        avg = total/4
        if avg >= 80:
            grade = "A"
        elif avg >= 60:
            grade = "B"
        elif avg >= 40:
            grade = "C"
        else:
            grade = "D"
        print(f"Grade: {grade}")
```

## Holiday Package Evaluator
### Goal:

Compare multiple travel packages across features like hotel, transport, food.

### Steps:

Outer loop ‚Üí 3 packages.

Inner loop ‚Üí 3 features.

Input rating out of 5.

Skip if rating = 0.

Print average rating per package.

### Example:
Package 1: [4, 5, 5] ‚Üí Avg = 4.67  
Package 2: [3, 4, 0] ‚Üí Avg = 3.5

```python
for pkg in range(1,4):
    print(f"Package {pkg}:")
    ratings = []
    for feat in range(1,4):
        rating = float(input(f"Feature {feat} rating (0-5): "))
        if rating == 0:
            continue
        ratings.append(rating)
    avg = sum(ratings)/len(ratings) if ratings else 0
    print(f"Average rating for Package {pkg}: {avg}")
```
